```python
#Computing molecular descriptors and fingerprints using RDKit
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose = False):
    moldata = []
    for elem in smiles:
        mol = Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData = np.arange(1,1)
    i = 0
    for mol in moldata:
        desc_MW = Descriptors.MolWt(mol)
        desc_molLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHacceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MW, desc_molLogP, desc_NumHDonors, desc_NumHacceptors])

        if i == 0:
            baseData = row
        else:
            baseData = np.vstack([baseData, row])

        i += 1
    columnsName = ['MW', 'LogP', 'NumHDonors', 'NumHacceptors']
    descriptors_df = pd.DataFrame(data=baseData, columns=columnsName)

    return descriptors_df
```

```python
df_lipinski = lipinski(df.canonical_smiles)
```

```python
df_lipinski.head(10)
```

```python
#Combining the two data together

df_combined = pd.concat([df3, df_lipinski], axis=1)
df_combined.head(10)
```

```python
def pIC50 (input):
    pIC50  = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9)
        pIC50.append(-np.log10(molar))
    input['pIC50'] = pIC50
    x = input.drop("standard_value_norm", axis=1)

    return x
```

```python
def norm_value(input):
    norm = []
    for i in input['standard_value']:
        try:
            val = float(i)
        except ValueError:
            val = np.nan
        if val > 10000000:
            val = 100000000
        norm.append(val)
    input['standard_value_norm'] = norm
    x = input.drop('standard_value', axis=1)
    return x
```

```python
df_norm = norm_value(df_combined)
```

```python
#Preparing SMILES Files
smiles_file = df3_class[['canonical_smiles', 'molecule_chembl_id']]
smiles_file.to_csv("smiles.smi", sep= '\t' ,index=False, header=False)
```

```python
#Calculating descriptors usinf Padel
import os
import subprocess

def run_padel_descriptor(
        input_smi = 'smiles.smi',
        output_csv = 'padel_output.csv',
        padel_jar = r'PaDEL-Descriptor\PaDEL-Descriptor.jar',
        fingerprints = True,
        threads = 4
):
    # check paths
    if not os.path.exists(input_smi):
        raise FileNotFoundError(f"SMILES file not found: {input_smi}")
    if not os.path.exists(padel_jar):
        raise FileNotFoundError(f"PaDEL jar not found: {padel_jar}")
    # construct command
    cmd = [
        "java", "-Xms2G", "-Xmx4G", "-jar", padel_jar,
        "-2d",
        "-dir", ".",
        "-file", output_csv,
        "-threads", str(threads)
    ]
    if fingerprints:
        cmd.append("-fingerprints")
    
    print("🔄 Running PaDEL-Descriptor... please wait (~2–5 min for 100–500 molecules)")
    subprocess.run(cmd, check=True)
    print(f"✅ PaDEL completed. Output saved to {output_csv}")
    
    # Load descriptors
    desc_df = pd.read_csv(output_csv)
    print(f"✅ Loaded {desc_df.shape[0]} molecules × {desc_df.shape[1]} descriptors")
    return desc_df
```

```python
#Running the descriptor dataframe
desc_df = run_padel_descriptor(
    input_smi="smiles.smi",
    output_csv= "Padel_output.csv",
    padel_jar= r"PaDEL-Descriptor\PaDEL-Descriptor.jar"
)
```

```python
# Displaying then descriptor dataframe
desc_df.head()
```

```python
#Merging the descriptor dataframe with the bioactivity data
merged_df = pd.merge(df3_class, desc_df, left_on='molecule_chembl_id', right_on='Name')
merged_df.to_csv('Final_bioactivity_descriptors.csv')
merged_df.head()
```

```python
#Sperrating the X (Independent value) and Y (Target column)
X = merged_df.drop(columns=['Name', 'molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'MW_x', 'LogP', 'NumHDonors', 'NumHacceptors', 'pIC50'])
Y_class = merged_df['bioactivity_class']
Y_reg = merged_df['pIC50']
X.shape
```

```python
#Filling up the missing values in X with 0
X.fillna(0, inplace=True)
```

```python
# Dropping highly correlated features
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.9)]
X = X.drop(columns=to_drop)
feature_names = X.columns
```

```python
# Now removing low variance features
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=0.16)
X = selection.fit_transform(X)
X.shape
```

```python
selected_features = feature_names[selection.get_support(indices=True)]
import pickle
with open('feature_columns.pkl', 'wb') as f:
    pickle.dump(selected_features, f)
    
```

```python
#Scaling the X values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
import pickle
pickle.dump(scaler, open('scalar.pkl', 'wb'))
```

```python
#Spliting dataste into training and testing set for regression
from sklearn.model_selection import train_test_split
X_train_reg, X_test_reg, Y_reg_train, Y_reg_test = train_test_split(X_scaled, Y_reg, test_size=0.2, random_state=42)

#splitting dataset into training and testing set for classification
X_train_class, X_test_class, Y_class_train, Y_class_test = train_test_split(X_scaled, Y_class, test_size=0.2, random_state=42)
```

```python
print(X_train_reg.shape)
print(X_test_reg.shape)
print(X_train_class.shape)
print(X_test_class.shape)
```

```python
# Comparing the regressor model
from lazypredict.Supervised import LazyRegressor

np.random.seed(100)
reg = LazyRegressor(verbose=0, ignore_warnings=True)
models, predictions = reg.fit(X_train_reg, X_test_reg, Y_reg_train, Y_reg_test)
models.sort_values(by="R-Squared", ascending=False).head(15)
```
